```python
# 1•	Assume a 30 x 30 grid. Plot these gridlines in matplotlib.
import numpy as np
import matplotlib.pyplot as plt
plt.title('30 x 30 grid')
for i in range(0,30):
    plt.plot(np.full((30, ), i), np.arange(0,30), color='orange')
    plt.plot(np.arange(0,30), np.full((30, ), i), color='orange')
#first I plotted columns and then rows.
```

```python
#2.•	Now take a shearing transformation matrix and apply that on that grid. Plot the new grid lines
shear=np.array([1,0,0.3,1]).reshape(2,2) #shear matrix for y shear
plt.title('Shear Transformation')
for i in range(0, 30):
    for j in range(0, 30):
        dot = shear.dot(np.array([i, j]).reshape(2, 1))
        plt.plot(dot[0], dot[1], 'g.')
#by multipyling shear matrix with grid we get vertical shear:
```

```python
#3•	Instead of doing the computation serially on 900 points during the 2nd step in loops, can you parallelize
# your approach to avoid loops?
matrix = np.empty((2,900))
index = 0

for i in range(-15, 15): # Looping x-coordinate of each point
    for j in range(-15, 15): # Looping y-coordinate of each point
        # Appending values
        matrix[0][index] = i
        matrix[1][index] = j
        index+=1
matrix = shear.dot(matrix)
plt.title('Shear Transformation')
for i in range(900):
    plt.plot(matrix[0][i], matrix[1][i], 'g.')
```
